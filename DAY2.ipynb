- 청불 -> 20
- 전체 관람가 -> 1
- 15세 관람가 -> 15
- 12세 관람가 -> 12

### def

```python
def change_rat(text:str):
  if text == '청소년 관람불가':
    return 20
  elif text == '15세 관람가':
    return 15
  elif text == '12세 관람가':
    return 12
  else:
    return 1

df['screening_rat'].map(change_rat)
```

### loc

```python
df.loc[df['screening_rat'] == '청소년 관람불가', 'screening_rat'] = 20
df.loc[df['screening_rat'] == '15세 관람가', 'screening_rat'] = 15
df.loc[df['screening_rat'] == '12세 관람가', 'screening_rat'] = 12
df.loc[df['screening_rat'] == '전체 관람가', 'screening_rat'] = 1
df
```

## 이전 작품 관객 수와 현재 작품 관객 수

```python
temp = df[df['dir_prev_bfnum'].notnull()]
temp
```

```python
plt.figure(dpi=150)

plt.title('이전 작품과 현재 작품의 관객 수')
plt.ylabel('관객 수')

plt.plot(temp['dir_prev_bfnum'], label='이전 작품', alpha=0.6)
plt.plot(temp['box_off_num'], label='현재 작품', alpha=0.6)

plt.legend()
plt.show()
```

```python
temp = temp.sort_values(by='box_off_num').reset_index(drop=True)

plt.figure(dpi=150)

plt.title('이전 작품과 현재 작품의 관객 수 by 현재 작품')
plt.ylabel('관객 수')

plt.plot(temp['dir_prev_bfnum'], label='이전 작품', alpha=0.6)
plt.plot(temp['box_off_num'], label='현재 작품', alpha=0.6)

plt.legend()
plt.show()
```

```python
temp = temp.sort_values(by='dir_prev_bfnum').reset_index(drop=True)

plt.figure(dpi=150)

plt.title('이전 작품과 현재 작품의 관객 수 by 이전 작품')
plt.ylabel('관객 수')

plt.plot(temp['dir_prev_bfnum'], label='이전 작품', alpha=0.6)
plt.plot(temp['box_off_num'], label='현재 작품', alpha=0.6)

plt.legend()
plt.show()
```
